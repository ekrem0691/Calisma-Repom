```python
sayı = int(input("Lütfen tablo ölçüsünü giriniz : "))
print(" ", end = " ")
for sütun in range(1, sayı + 1):
    print("{0:4}".format(sütun), end = " ")
print()
print("+", end = " ")
for sütun in range(1, sayı + 1):
    print("-----", end = "")
print()
for satir in range(1, sayı + 1):
    print("{0:3}|".format(satir), end = " ")
    for sütun in range(1, sayı + 1):
        değer = satir * sütun
        print("{0:4}".format(değer), end = " ")
    print()
```

```python
girdi = input("Lütfen permütasyonunu istediğiniz kelimeyi yazınız :")
listem = []
for ilk in girdi:
    for ikinci in girdi:
        if ikinci != ilk:
            for üçüncü in girdi:
                if üçüncü != ilk and üçüncü != ikinci:
                    for dördüncü in girdi:
                        if dördüncü != ilk and dördüncü != ikinci and dördüncü != üçüncü:
                            listem.append(ilk + ikinci + üçüncü + dördüncü)
print(listem)
```

Y/y yazınca devam eden ; N/n yzınca sonlanan bir döngü: kaç kere döndüğü yazılacak

```python
sayac = 0
while True:
    giriş = input("Devam etmek için 'Y/y'; çıkmak için 'N/n' yazınız!")
    if giriş == "y" or giriş == "Y":
        sayac += 1
        print(sayac)
    elif giriş == "N" or giriş == "n":
        break
    elif giriş != "N" or giriş != "n":
        print("{} geçerli değildir.".format(giriş))
print(sayac)
```

```python
yükseklik = int(input())
satır = 0

while satır < yükseklik:
    sayac = 0
    while sayac < yükseklik - satır:
        print(end = " ")
        sayac += 1
    sayac += 0
    while sayac < 2 * satır + 1:
        print(end = "*")
        sayac += 1
    print()
    satır += 1
    
```
